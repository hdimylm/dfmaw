Perform the following operations on iris dataset
1. List down the features and their types
2. Create a box plot  and histogram for each feature in the dataset.
3. Compare distributions and identify outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Iris.csv')
```

```python
#1
df.dtypes
```

```python
#2
for column in df.columns:
     plt.figure(figsize=(10, 6))
     sns.boxplot(x=df[column])
     plt.title(f'Box plot of {column}')
     plt.show()
```

```python
for column in df.columns:
  plt.figure(figsize=(10, 6))
  sns.histplot(df[column], bins=20, kde=True)
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

```python
#3
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.title('Boxplot of SepalLengthCm, SepalWidthCm, PetalLengthCm and PetalWidthCm')
plt.show()
```
